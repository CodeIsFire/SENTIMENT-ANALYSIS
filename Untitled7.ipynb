```python
#IMPORTING THE NECESSARY LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
import io
# Load dataset
from google.colab import files
uploaded=files.upload()
df = pd.read_csv(io.BytesIO(uploaded['YoutubeCommentsDataSet.csv']))
df
```

```python
#IMPORTING RE FOR DATA CLEANING BY SUBBING
import re
def clean_text(text):
    if isinstance(text, str):
        text = re.sub(r'http\S+', '', text)
        text = re.sub(r'@\w+', '', text)
        text = re.sub(r'[^\w\s]', '', text)
        text = text.lower()
        return text
    return ''

df['cleaned_comment'] = df['Comment'].apply(clean_text)

display(pd.DataFrame({
    'Original Comment': df['Comment'].head(),
    'Cleaned Comment': df['cleaned_comment'].head()
}))
```

```python
#Sentiment Analysis with TextBlob
def analyze_sentiment(text):
    analysis = TextBlob(text)
    return pd.Series([analysis.sentiment.polarity, analysis.sentiment.subjectivity])

df[['polarity', 'subjectivity']] = df['cleaned_comment'].apply(analyze_sentiment)

display(df[['Comment', 'polarity', 'subjectivity']].head())
```

```python

def sentiment_category(polarity):
    if polarity > 0.2:
        return 'Positive'
    elif polarity < -0.2:
        return 'Negative'
    else:
        return 'Neutral'

def subjectivity_category(subjectivity):
    if subjectivity > 0.6:
        return 'Highly Subjective'
    elif subjectivity < 0.3:
        return 'Mostly Objective'
    else:
        return 'Moderate'

df['sentiment'] = df['polarity'].apply(sentiment_category)
df['subjectivity_cat'] = df['subjectivity'].apply(subjectivity_category)

display(df[['Comment', 'polarity', 'sentiment', 'subjectivity', 'subjectivity_cat']].head())
```

```python
fig = plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.histplot(df['polarity'], kde=True, bins=30, color='skyblue')
plt.title('Sentiment Polarity Distribution')
plt.xlabel('Polarity')
plt.axvline(x=0, color='red', linestyle='--')

plt.subplot(2, 2, 2)
sns.histplot(df['subjectivity'], kde=True, bins=30, color='lightgreen')
plt.title('Subjectivity Distribution')
plt.xlabel('Subjectivity')

plt.subplot(2, 2, 3)
sentiment_counts = df['sentiment'].value_counts()
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'lightyellow', 'lightcoral'])
plt.title('Sentiment Categories')

plt.subplot(2, 2, 4)
sns.countplot(y='subjectivity_cat', data=df, palette='viridis')
plt.title('Subjectivity Categories')
plt.tight_layout()

plt.show()
```
